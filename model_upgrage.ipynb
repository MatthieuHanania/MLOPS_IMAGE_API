### this notebook is used to upgrade the model on the new data

```python
#TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras
from datetime import datetime

#Helper libraries
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

```python
df_mnist = pd.read_csv("fashion-mnist-train-2.csv")
Y_train = df_mnist.label
X_train = df_mnist.drop("label",axis=1)

X_train=X_train/255
```

```python
model = keras.models.load_model('mnistModel.h5')
model.fit(X_train,Y_train,epochs=100)
model.save('mnistModel.h5')
```
